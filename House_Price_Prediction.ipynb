```python
from utils import layer_sizes,initialize_parameters,forward_propagation,backward_propagation,cost_function,update_parameters,predict_value,neural_network_model
import numpy as np
import matplotlib.pyplot as plt
# A library for data manipulation and analysis.
import pandas as pd

# Output of plotting commands is displayed inline within the Jupyter notebook.
%matplotlib inline
#for consistent result of random numbers
np.random.seed(3)
```

```python
df = pd.read_csv("C:/pythonprojects/calculas_neural_network/datas/house_prices_train.csv")
X=df[['GrLivArea', 'OverallQual']]  #here we are selecting the ground living area  and overall quality based on which we are making the salesprice prediction
X
```

```python
Y=df[['SalePrice']]
Y
```

```python
#normalizing the data

X_norm = (X -np.mean(X,axis = 0)) / np.std(X,axis = 0)
X_norm
Y_norm = (Y -np.mean(Y,axis = 0)) / np.std(Y,axis = 0) 
X_norm
```

```python
X_multiple = np.array(X_norm).T #so here we are converting the input data into numpy array having feature in row and samples of each feature as columns
X_multiple
```

```python
Y_multiple = np.array(Y_norm).reshape((1,len(Y_norm)))  #the original sales value converted into numpy array having one row and multiple columns 
Y_multiple
```

```python
#defining the neural network structure


(n_x, n_y) = layer_sizes(X_multiple, Y_multiple)
print("The size of the input layer is: n_x = " + str(n_x))
print("The size of the output layer is: n_y = " + str(n_y))
```

```python

parameters = initialize_parameters(n_x,n_y)
parameters
```

```python

Y_predicted = forward_propagation(X_multiple,parameters)
Y_predicted  #the predicted sale price initially
```

```python
#calculating the cost function
cost_function(Y_predicted,Y_multiple)
```

```python

grads = backward_propagation(Y_predicted,Y_multiple,X_multiple)
grads

```

```python

updated_paramters = update_parameters(parameters,grads,0.1)
updated_paramters

```

```python

final_parameters = neural_network_model(X_multiple,Y_multiple,30,0.1,print_cost=False)

```

```python
X_pred = np.array([[1716,7],[1262,6,],[1786,7]]).T  #here we are passing the pair of samples of two features

Y_pred = predict_value(df[['GrLivArea', 'OverallQual']],df['SalePrice'],X_pred,final_parameters)
Y_pred
```
